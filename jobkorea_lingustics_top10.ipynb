BeautifulSoup를 메모리에 올리고, 

```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

잡코리아의 url을 가져왔다. 

```python
html = urlopen('http://www.jobkorea.co.kr/Top100/?Main_Career_Type=1&Search_Type=2&BizJobtype_Bctgr_Code=0&BizJobtype_Bctgr_Name=%EC%A0%84%EC%B2%B4&Major_Big_Code=1&Major_Big_Name=%EC%96%B4%EB%AC%B8%ED%95%99&Edu_Level_Code=9&Edu_Level_Name=%EC%A0%84%EC%B2%B4')
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

10대 기업의 이름 목록은 <b> </b>사이에 있다는 것을 알고, b를 가져온 후 

```python
c =  bsObject.body.find_all('b')
c
```

<b>의 내용만을 가져왔다. 
이전에 'content' 만을 가져오려 했으나, 이상하게 오류가 떴다. 
교수님의 학생 한 분의 이메일에 답장한 바와 같이, get_text, find_all, select 등의 차이점이 있는 것 같다.   

```python
for cn in bsObject.find_all('b'):
    print(cn.get_text())
    
```

마감일자를 찾기 위해서 <span>이라는 태그 속 day를 사용한다는 것을 알았으나, 
    아래 3가지 방법 모두 성공하지 못함. 

```python
for span in bsObject.find_all('span'):
    print(span.get_text('day'))
```

```python
for link in bsObject.find_all('span'):
    for dday in link.get_text('day'):
        print(dday[-1])
```

```python
for link in bsObject.find_all('span'):
    print(link.get('content'))
```
